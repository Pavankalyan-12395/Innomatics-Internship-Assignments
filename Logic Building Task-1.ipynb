### 1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

```python
username = "admin"
password = "1234"
user = input("Enter username: ")
pwd = input("Enter password: ")
if user == username and pwd == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

```python
marks = [45, 78, 90, 33, 60]
#assigning initial count of Pass/Fail students =0
pass_count = 0
fail_count = 0
# To find pass/fail of all students using for loop
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

### 3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = [] #definig empty list for futher purpose
for name in names:
    cleaned = name.strip().lower()   # remove spaces + convert to lowercase
    cleaned_names.append(cleaned)
print("Cleaned Names:", cleaned_names)
```

### 4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' , Length: {length}")
    if length > 10:
        print("→ This message is LONG\n")
    else:
        print("→ This message is SHORT\n")
```

### 5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for entry in logs:
    if entry == "ERROR":
        error_count += 1
print("Total ERROR entries:", error_count)
```
