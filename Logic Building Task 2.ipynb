### Problem 1: Unique Words in a Sentence
##### Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves
##### Hint
Split the sentence into individual words.
Convert the list of words into a set to remove duplicates.
Print both the set and its length.
##### Input
Enter sentence: python is easy and python is powerful
##### Expected Output
Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}

```python
#Get input from the user
sentence = input("Enter sentence: ")
#Split the sentence into a list of words
words = sentence.lower().split()
#Convert the list to a set to filter out duplicates
unique_words = set(words)
#Display the results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

### Problem 2: Highest Salary from Employee Data
##### Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.
##### Hint
Use a dictionary where the key is the employee name and the value is the salary.
Compare salary values to find the highest.
##### Input Format
A dictionary where:
Keys represent employee names
Values represent employee salaries
##### Example Input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}


```python
# Step 1: Given the employee data
employees = {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
# Step 2: Use the max() function with a "key" argument
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
# Step 3: Display the output
print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

### Problem 4: Count Products Above a Price Threshold
##### Problem Statement
Count how many products have a price greater than 1000.
##### Hint
Use a loop to compare each price with 1000.
##### Input Format
A list of integers where each element represents the price of a product.
##### Example Input
prices = [450, 1200, 899, 1500, 300]


```python
# Define the list of product prices
prices = [450, 1200, 899, 1500, 300]
# Initialize a counter at zero
count = 0
# Loop through each price in the list
for price in prices:
    # Check if the price is greater than 1000
    if price > 1000:
        # Increment the counter if the condition is met
        count += 1
# Display the result
print(f"Products above 1000: {count}")
```

### Problem 7: Count Character Frequency
##### Problem Statement
Count how many times each character appears in a given string.
##### Hint
Use a dictionary.
Increment the count for each character occurrence.


```python
# Define the input string
text = "pythonp"
# Initialize an empty dictionary to store counts
char_counts = {}
# Loop through each character in the string
for char in text:
    if char in char_counts:
        # If character is already a key, increment its value by 1
        char_counts[char] += 1
    else:
        # If character is not in dictionary, add it with a value of 1
        char_counts[char] = 1
# Step 4: Display the result
print(char_counts)
```

### Problem 10: Calculate Average Marks
##### Problem Statement
Calculate the average marks of a student from a list of marks.
##### Hint
Calculate the sum of all values.
Divide by the total number of marks.

```python
# Define the list of marks
marks = [85, 62, 78, 90, 47]
# Calculate the sum of all marks
total_sum = sum(marks)
# Count the total number of subjects
total_subjects = len(marks)
# Calculate the average
average = total_sum / total_subjects
# Display the result
print(f"Average Marks: {average}")
```
