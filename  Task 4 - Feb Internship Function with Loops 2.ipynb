### Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry


```python
# Function to validate recharge
def validate_recharge(amount, valid_plans):
    # Check minimum amount condition
    if amount >= 50:
        # Check if it is one of the valid plan amounts
        if amount in valid_plans:
            return True
        else:
            print("Invalid plan! Available plans are:", valid_plans)
            return False
    else:
        print("Recharge amount must be at least ₹50.")
        return False


# Main Program
valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        # Call validation function
        if validate_recharge(amount, valid_plans):
            print(f"Recharge of ₹{amount} successful!")
            break
        else:
            print("Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```

### Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary

```python
# Function to check stock level
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Dictionary of products and stock quantities
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Milk": 12,
    "Rice": 40,
    "Oil": 8
}

# Call function
check_inventory(inventory)
```

###  Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
# Function to process student results
def process_result(marks):
    total = 0
    
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    
    # Calculate average
    average = total / len(marks)

    # Print result based on average
    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"


# Example list of marks
student_marks = [45, 60, 55, 40, 70]

# Call function and print result
print(process_result(student_marks))
```

### Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry


```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12

    # Normal fare calculation
    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra during peak hours
    if peak_hour:
        fare *= 1.25

    return fare


# Main Program with retry option
while True:
    try:
        distance = float(input("Enter distance travelled (km): "))
        peak = input("Is it peak hour? (yes/no): ").strip().lower()

        if peak == "yes":
            peak_hour = True
        elif peak == "n0":
            peak_hour = False
        else:
            print("Invalid input for peak hour!\n")
            continue

        # Calculate fare
        total_fare = calculate_fare(distance, peak_hour)
        print(f"Total Cab Fare: ₹{total_fare:.2f}")

        # Ask if user wants to retry
        retry = input("Do you want to calculate again? (yes/no): ").strip().lower()
        if retry != "yes":
            print("Thank you! Have a safe ride.")
            break
        print()

    except ValueError:
        print("Invalid distance! Please enter a number.\n")
```

### Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement


```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    present_count = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1

    # Calculate percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    # Return result
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"


# Example input
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function and print result
print(check_attendance(attendance))
```

### Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password: Must be at least 8 characters long."

    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Check all conditions
    if has_digit and has_special:
        return "Strong Password"
    elif not has_digit:
        return "Weak Password: Must include at least one digit."
    elif not has_special:
        return "Weak Password: Must include at least one special character (@, #, $)."


# Example
password = "hello@123"
print(check_password(password))
```
